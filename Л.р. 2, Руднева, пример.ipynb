```
#Построить столбиковую диаграмму по возрасту и соответствующему их количеству
import numpy as np
import matplotlib.pyplot as plt 
import seaborn

import pandas as pd
data = pd.DataFrame
data = pd.read_csv('titanic_train.csv')
data = data.sort_values(by='age')
groupby_age = data.groupby('age').size() 
groupby_age.plot.bar(title = 'Distribution of age', figsize = (15,4)) 
plt.show()
```

```
#Построить столбиковую диаграмму по половому признаку и соответствующему их количеству
import numpy as np
import matplotlib.pyplot as plt 
import seaborn

import pandas as pd
data = pd.DataFrame
data = pd.read_csv('titanic_train.csv')
data = data.sort_values(by='sex')
groupby_age = data.groupby('sex').size() 
groupby_age.plot.bar(title = 'Distribution of sex', figsize = (15,4)) 
plt.show()
```
